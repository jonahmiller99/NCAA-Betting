```python
# Putting it All Together
# Team_DataFrame is a dictionary of dataframes containing the top 80 teams from the 2019 season and their first 30
# games and stats

d = {}
for j in range(28):
    d[j] = Col_Names[j]


for i in range(len(teamwork)):
    df = pd.DataFrame(teamwork[teams[i]])
    l = [4,5,6,12,32]
    df = df.drop([4,5,6,12,32], axis=1)
    df.columns = range(df.shape[1])
    df = df.rename(d, axis=1)
    Team_DataFrame[teams[i]] = df
    
    
for j in range(len(teamwork)):
    final_spread = []
    df = Team_DataFrame[teams[j]]
    for i in range(len(df)):
        x = df['Result'][i]
        cat = x.split(',')
        win_loss = cat[0]
        score = cat[1]
        differential = 1
        if win_loss == 'W':
            differential = 1
        else:
            differential = -1
            
        split = score.split("-")
        same = True
        if '2OT' in split[1]:
            split[1] = split[1].replace('2OT', '')
            same = False
        if '3OT' in split[1] and same:
            split[1] = split[1].replace('3OT', '')
            same = False
        if 'OT' in split[1]:
            split[1] = split[1].replace('OT', '')
            same = False
        difference = int(split[0]) - int(split[1])
        end_result = difference * differential 
        final_spread.append(end_result)
    df['Final Spread'] = final_spread

for k in range(len(teamwork)):    
    Cover = []
    df = Team_DataFrame[teams[j]]
    for i in range(len(df)):
        s = df['Spread'][i]
        t1 = (df['Team'][i])
        t2 = (df['Opponent'][i])
        t1 = t1.replace("+", " ")
        t2 = t2.replace("+", " ")
        seperated_spread = []
        favored_team = False
        if t1 in s:
            favored_team = True
            seperated_spread = s.split(t1)
            score = seperated_spread[1].split(',')[0]
        else:
            seperated_spread = s.split(t2)
 
            score = seperated_spread[1].split(',')[0]
        if favored_team:
            if int(df['Final Spread'][i]) > -1*float(score):
                Cover.append(1)
            else:
                Cover.append(0)
        else:
            if float(score) < int(df['Final Spread'][i]):
                Cover.append(1)
            else:
                Cover.append(0)
    df['Cover'] = Cover

    
```
